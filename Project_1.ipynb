```python
## 1. Load the Data.
import pandas as pd
data = pd.read_csv("C:/Users/ASUS/OneDrive/Desktop/Internship/retail_sales_dataset.csv")
print(data)
```

```python
## 2. Inspect the Data.
print(data.head(10))
print(data.tail(10))
```

```python
## 3(a). Basic Function to Understand about Data.
print(data.info())
```

```python
#3(b). 
print(data.describe())
```

```python
## 4.  Check for Missing Value.
print(data.isnull().sum())
```

```python
## 5. Check For Duplicate Value.
print(data["Transaction ID"].duplicated().sum())
```

```python
## Since there is no null and duplicate value so we can go further.
```

```python
## 6. Statistics Data
# 1. Mean.
print(data["Price per Unit"].mean())
print(data["Total Amount"].mean())
```

```python
# 2. Median.
print(data["Price per Unit"].median())
print(data["Total Amount"].median())
```

```python
# 3. Standard Deviation.
print(data["Price per Unit"].std())
print(data["Total Amount"].std())
```

```python
# 4.  Mode Function
print(data["Price per Unit"].mode())
print(data["Total Amount"].mode())
```

```python
## 7. Converting Date to DateTime Column.
data["Date"] = pd.to_datetime(data["Date"])
print(data["Date"].dtype)
print(data)
```

```python
## 8.  Detect Seasonbility.
# 1. Based on Month.
data["Month"] = data["Date"].apply(lambda x: x.month)
print(data)
```

```python
## 2.  Based on Year.
data["Year"] = data["Date"].apply(lambda x :x.year)
print(data)
```

```python
## Bar Plot Based on Gender and Sum of Total Amount.
import matplotlib.pyplot as plt
gp = data.groupby("Gender")["Total Amount"].sum()
print(gp)
colors = ["brown","green"]
plt.bar(gp.index,gp.values,edgecolor = "black",color = colors)
plt.xlabel("Gender")
plt.ylabel("Total Amount")
plt.title("Total Amount Spent Based on Gender")
plt.show()
```

```python
## Bar Plot Based on Gender and Total Quantity of Product.
import matplotlib.pyplot as plt
gp = data.groupby("Gender")["Quantity"].sum()
print(gp)
colors = ["pink","orange"]
plt.bar(gp.index,gp.values,edgecolor = "black",color = colors)
plt.xlabel("Gender")
plt.ylabel("Product Quantity")
plt.title("Total number of products purchased by customers, categorized by gender.")
plt.show()
```

```python
## Average Amount spent by Customer based on Age using line Plot.
import matplotlib.pyplot as plt
gp = data.groupby("Age")["Total Amount"].mean()
print(gp)
plt.plot(gp.index,gp.values,ls = "-.",marker = "D",color = "blue")
plt.xlabel("Age")
plt.ylabel("Average Total Amount")
plt.title("Average Amount spent by Customer based on Age")
plt.show()
```

```python
## Pie Plot Based on Category.
import matplotlib.pyplot as plt
gp = data.groupby("Product Category")["Total Amount"].sum()
print(gp)
ex = [0.05,0,0]
plt.pie(gp.values,labels = gp.index,autopct = "%0.2f",explode = ex,shadow = True,colors = ["red","brown","green"])
plt.title("Pie Chart Based on Category")
plt.show()
```

```python
## Total Amount spent based on Month.
import seaborn as sns
import matplotlib.pyplot as plt
gp = data.groupby("Month").agg({"Total Amount":"sum"})
sns.heatmap(gp)
print(gp)
```

```python
## Scatter Plot of Quantity and Total Amount
import matplotlib.pyplot as plt
import numpy as np
gp = data.groupby("Quantity")["Total Amount"].sum()
print(gp)
color = np.random.randint(1,50,4)
plt.scatter(gp.index,gp.values,marker = "D",c = color,cmap = "rainbow",edgecolors = "black")
plt.xlabel("Quantity")
plt.ylabel(" Sum of Total Amount")
plt.title("Sum of Total Amount Based On Quantity")
plt.colorbar()
plt.show()
```
