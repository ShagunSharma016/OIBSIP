```python
## Total Spend.
import matplotlib.pyplot as plt
data["Total SPend"]  =data[['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts',	'MntGoldProds','MntRegularProds']].sum(axis=1)
# print(data)
df = data.groupby("Income").agg({"Total SPend":"sum"})
# print(df)
plt.scatter(df.index,df.values,edgecolor = 'black',color = 'red',)
plt.title("Income Vs Total Spend",fontsize = 20)
plt.show()
```

```python
## Droping the Rows if the Incomes is less than Total Spend.
data.drop(data[data['Income']<data['Total SPend']].index,inplace = True)
print(data)
```

```python
## Using Quartile Function to Detect Outliers.
Max_qurt1 = data["Income"].quantile(0.999);
Min_qurt1 = data["Income"].quantile(0.001);
Max_qurt2 = data["Total SPend"].quantile(0.999);
Min_qurt2 = data["Total SPend"].quantile(0.001)
print(Max_qurt1,"-",Min_qurt1,"-",Max_qurt2,"-",Min_qurt2)
```

```python
data.drop(data[(data["Income"]>=Max_qurt1) | (data["Income"]<=Min_qurt1) | (data["Total SPend"]>=Max_qurt1) | (data["Total SPend"]<=Min_qurt2)].index,inplace = True)
print(data)
```
