```python
## 1. Load the DataSet.
import pandas as pd
data = pd.read_csv("C:/Users/ASUS/OneDrive/Desktop/Internship/AB_NYC_2019.csv")
print(data)
```

```python
## 2. Inspect the Data.
# a. Head Function.
print(data.head(10))
```

```python
# b. Tail Function.
print(data.tail(10))
```

```python
# c. Info Function.
print(data.info())
```

```python
# d. Describe Function.
print(data.describe())
```

```python
## 3. Working With Duplicate Value.
# a. Checking for Duplicate value in Important column.
print(data["id"].duplicated().sum())
print(data["host_id"].duplicated().sum())
```

```python
# b.Analize the duplicate data.
duplicate_data = data[data["host_id"].duplicated()]
print(duplicate_data)
```

```python
# c. Drop the Duplicate Value.
data.drop_duplicates(subset = ["host_id"],keep = "first",inplace = True)
print(data["host_id"].duplicated().sum())
print(data)
```

```python
## 4. Working with Null Value.
# a. Checking null value in entire dataset.
print(data.isnull().sum())
```

```python
# b. Replacing null value of name column to Unknown name.
data["name"].fillna("Unknown_name",inplace = True)
print(data)
```

```python
# c. Replacing Null value of host name column to unknown host.
data["host_name"].fillna("Unknown_host",inplace = True)
print(data)
```

```python
# d. Replacing Null value of reviews_per_month  column with 0.
data["reviews_per_month"].fillna(0,inplace = True)
print(data)
```

```python
# e. Replacing null value of last_review with 0.
data["last_review"].fillna(0,inplace = True)
print(data.isnull().sum())
print(data)
```

```python
## 5. Detect Outliers.
# a. Using Box Plot.
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(data = data,x = "room_type",y = "price")
plt.show()
```

```python
# b. IQr and Clip Function.
Q1 = data["price"].quantile(0.25)
print(Q1)

Q3 = data["price"].quantile(0.75)
print(Q3)

IQR = Q3-Q1
print(IQR)

upper_limit = Q3 + 1.5*IQR
lower_limit = Q1 -1.5*IQR
print(f"Upper limit : {upper_limit}")
print(f"Lower limit : {lower_limit}")
data_capped = data.copy()
data_capped["price"] = data_capped["price"].clip(upper = None,lower = lower_limit)
data_capped["price"] = data_capped["price"].clip(upper = upper_limit)
print(data_capped)
```

```python
# c. Outlires Removed.
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(data = data_capped,x = "room_type",y = "price")
plt.show()
```
